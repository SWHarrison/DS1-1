```python
## Data gathering and some clean-up
import pandas as pd
import seaborn as sns
sns.set(rc={'figure.figsize':(22.7,16.27)})
locations = ['LA','NY','SF','SG','SV','HK', 'Taipei', 'Tokyo']
df = pd.DataFrame()
for i in range (len(locations)):
    for week in range (1,8):
        file_name = "./2016/Anon Week " + str(week) + " Feedback - " + str(locations[i]) + ".csv"
        #print(file_name)
        try:
            new_df = pd.read_csv(file_name)
            new_df["Location"] = locations[i]
            new_df['Week'] = week
            new_df['Year'] = 2016
            df = pd.concat([df,new_df], ignore_index=True)
        except (FileNotFoundError):
            print(file_name)
    
df = df.drop('Unnamed: 0', axis = 1)
df = df.rename(index=str,columns = {"How well are the tutorials paced?":"tutorial pacing","How well is the schedule paced?": "schedule pacing", "How would you rate your overall satisfaction with the Summer Academy this week?": "weekly satisfaction", "What track are you in?" : "Track" })    

df2016 = pd.read_csv("./2016/Week 8 Feedback (2016, incomplete) - results.csv")
df2017 = pd.read_csv("./2017/Student Feedback Surveys-Superview.csv")
df2016 = df2016.rename(index=str, columns = {"How likely is it that you would recommend the Make School Summer Academy to a friend?": "Rating", "track" : "Track"})
df2017 = df2017.rename(index=str, columns = {"Rating (Num)" : 'Rating'})
df2016["Year"] = 2016
df2016["Week"] = 8
df2017["Year"] = 2017
```

```python
def _rating(data):
    print(data)
    print(type(data))
    try:
            return int(int(data)*2)
    except ValueError:
        return 0

df["Rating"] = df["weekly satisfaction"].apply(_rating)
df[df.Week == 3]
```

```python
def _remove_header(data):
    return int(str(data)[5])

df2017["Week"] = df2017["Week"].apply(_remove_header)

df2017
```

```python
## Calculating NPS score for 2016
NPS_promoters = len(df2016[df2016["Rating"] > 8])
NPS_passives = len(df2016[df2016["Rating"] > 6])
NPS_detractors = len(df2016[df2016["Rating"] < 7])
(NPS_promoters - NPS_detractors) / (len(df2016))
```

```python
## Average ratings for end of 2016, week 8
df2016["Rating"].value_counts().plot(kind='bar')
```

```python
## NPS score for 2017
NPS2017_promoters = len(df2017[df2017["Rating"] > 8])
NPS2017_passives = len(df2017[df2017["Rating"] > 6])
NPS2017_detractors = len(df2017[df2017["Rating"] < 7])
(NPS2017_promoters - NPS2017_detractors) / (len(df2017))
```
